```python
def pos_nodes_for_SNR(pos_nodes):
  return np.repeat(np.array(pos_nodes)[None, ...], len_snr, 0)

def pos_antenna_nodes_for_SNR(pos_nodes, no_antenna):
  return np.repeat(pos_nodes_for_SNR(pos_nodes)[..., None], no_antenna, -1)
def path_loss(pos_1, pos_2):
#     trick to put dim of pos_1 to axis 1 and pos_2 to axis 2
    pos_1_repeated = pos_nodes_for_SNR(pos_1)
    pos_2_repeated = pos_nodes_for_SNR(pos_2)
    
    pos_1_repeated = np.expand_dims(pos_1_repeated, -1)
    pos_2_repeated = np.expand_dims(pos_2_repeated, 1)
    
    dist = abs(pos_1_repeated-pos_2_repeated)
#     dist = LA.norm(pos_1_repeated-pos_2_repeated)
    
    pl = np.power(dist, -eff_path)*(10**(-3))
#     print(pl)    
    return pl
 
  
  
def get_channel(pos_1, pos_2,len_snr):
    pl = path_loss(pos_1, pos_2)
#     print('pl',pl.shape)
    pos_1 = np.array(pos_1)
    pos_2 = np.array(pos_2)
    
    channel_size = (len_snr, pos_1.shape[0], pos_2.shape[0])
    h = sqrt(pl/2)* (randn(*channel_size)+randn(*channel_size)*1j)
    
   
    return h
  
def get_channel_whole(pos_1, pos_2, no_sample = no_sample):
#     repeat the process to get the same channel length of no_sample
   channels = np.concatenate([get_channel(pos_1, pos_2, len_snr)[None]
                               for i in range(no_sample)])
   return channels

  
def get_channel_antenna(pos_1, antenna_1, pos_2, antenna_2,len_snr):
    pl = path_loss(pos_1, pos_2)
#     print('pl',pl.shape)
    pos_1 = np.array(pos_1)
    pos_2 = np.array(pos_2)
    
    pos_1 = pos_antenna_nodes_for_SNR(pos_1, antenna_1)
    pos_2 = pos_antenna_nodes_for_SNR(pos_2, antenna_2)
    channel_size = (len_snr, pos_1.shape[-1], pos_2.shape[-1])
 
    h = np.sqrt(pl/2) * (randn(*channel_size)+randn(*channel_size)*1j)
    #channel = np.power(abs(h), 2)
    
    return h  
  
def get_channel_antenna_whole(pos_1,antenna_1, pos_2, antenna_2, no_sample = no_sample):
#     repeat the process to get the same channel length of no_sample
   channels = np.concatenate([get_channel_antenna(pos_1,antenna_1, pos_2, antenna_2,len_snr)[None,...]
                               for i in range(no_sample)])
   return channels
  
def get_channel_antenna_whole_ofdma(pos_1,antenna_1, pos_2, antenna_2, no_sample = no_sample):
#     repeat the process to get the same channel length of no_sample
   
   channels = np.concatenate([np.concatenate([get_channel_antenna(pos_1,antenna_1, pos_2, antenna_2,len_snr)[None,...]
                               for i in range(no_sample)]) for j in range(no_subcarriers) ])
   return channels  
```
